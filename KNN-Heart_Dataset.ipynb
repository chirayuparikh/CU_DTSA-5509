The document outlines a step-by-step process for implementing a K-Nearest Neighbors (KNN) algorithm to predict diabetes using a dataset. Here's a summary of each step:

    Data Loading: The data is loaded into a Pandas DataFrame from a CSV file named diabetes.csv. Initial exploration includes checking the first few rows and the data structure using .head() and .info() methods.

    Data Preprocessing:
        Checked for null values in the dataset.
        Replaced zero values in 'SkinThickness', 'Insulin', 'Glucose', 'BloodPressure', and 'BMI' columns with their respective medians or means as zeros in these columns are not realistic.
        Removed outliers from several columns using the Interquartile Range (IQR) method to ensure data quality.
        Cleaned data is checked again for null values and confirmed clean.

    Exploratory Data Analysis (EDA):
        Visualized the correlation between different variables using a heatmap.
        Displayed frequency counts of the outcome variable using a bar chart.
        Plotted scatter and box plots to observe relationships and distributions involving 'Glucose', 'BMI', and 'Age' in relation to the diabetes outcome.

    Feature Selection and Data Splitting:
        Extracted features (independent variables) and outcomes (dependent variable) into separate variables.
        Split the data into training (80%) and testing (20%) sets.

    Data Normalization:
        Applied scaling on the feature set using StandardScaler to standardize the data, which is crucial for distance-based algorithms like KNN.

    Model Training and Prediction:
        Trained the KNN model with 3 neighbors.
        Made predictions on the test dataset.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score,f1_score
```

```python
df = pd.read_csv('diabetes.csv')
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
# Correct 0 values in SkinThickness

df['SkinThickness'] = df['SkinThickness'].replace(0,df['SkinThickness'].median())
# Correct 0 values in Insulin and Glucose
df['Insulin'] = df['Insulin'].replace(0, df['Insulin'].median())
df['Glucose'] = df['Glucose'].replace(0,df['Glucose'].mean())
# Correct 0 values in BloodPressure
df['BloodPressure'] =df['BloodPressure'].replace(0,df['BloodPressure'].mean())
df['BMI'] =df['BMI'].replace(0,df['BMI'].median())
```

```python
def detect_outliers(data,cols):
    for x in cols:
        q75,q25 = np.percentile(data.loc[:,x],[75,25])
        intr_qr = q75-q25

        max = q75+(1.5*intr_qr)
        min = q25-(1.5*intr_qr)

        data.loc[data[x] < min,x] = np.nan
        data.loc[data[x] > max,x] = np.nan
```

```python
outliers_drop = detect_outliers(df, ["Pregnancies", 'Glucose', 'BloodPressure', 'BMI', 
                                                      'DiabetesPedigreeFunction', 'SkinThickness', 
                                                      'Insulin', 'Age'])
```

```python
df = df.dropna(axis = 0)
df.isnull().sum()
```

```python
X = df.drop(["Outcome"], axis=1)
print(X)
y = df["Outcome"]
print(y)
```

```python
# train test split
X_train, X_test, y_train, y_test =train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
# Scale the features using StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
y_pred
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
knn.score(X_test,y_test)
```

```python
knn.score(X_train,y_train)
```
